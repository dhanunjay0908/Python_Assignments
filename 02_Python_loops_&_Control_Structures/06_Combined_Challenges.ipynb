##### Problem 6.1: Prime Numbers in Range (Medium)

Write a program to print all prime numbers between two given numbers (inclusive).

Example: Between 10 and 30: 11, 13, 17, 19, 23, 29

**Hint:** Outer loop for each number in range. For each number, use inner loop to check if it's prime (like Problem 2.5).

````python
while True:
    try:
        n1 = int(input('Enter 1st number: '))
        n2 = int(input('Enter 2nd number: '))
        break
    except ValueError:
        print('Enter a valid number.')
        
for i in range(n1, n2+1):
    count = 0
    for j in range(1, i+1):
        if i%j == 0:
            count+=1
    if count == 2:
        print(i)
````

##### Problem 6.2: GCD Calculator (Medium)

Write a program to find the Greatest Common Divisor (GCD) of two numbers using the Euclidean algorithm.

Algorithm:
1. Divide the larger number by the smaller number
2. Replace the larger number with the smaller number
3. Replace the smaller number with the remainder
4. Repeat until the remainder is 0
5. The GCD is the last non-zero remainder

**Hint:** While b != 0: store remainder (a%b), then a=b, b=remainder. When loop ends, a is the GCD.

````python
# import math
# n1 = int(input())
# n2 = int(input())
# print(math.gcd(n1, n2)) => using built-in method

while True:
    try:
        n1 = int(input('Enter 1st number: '))
        n2 = int(input('Enter 2nd number: '))
        break
    except ValueError:
        print('Enter a valid number.')
        
def gcd(n1, n2):
    while n2!=0:
        r = n1%n2
        n1 = n2
        n2 =  r
    return n1
print(gcd(n1, n2))
````

##### Problem 6.3: Perfect Number Checker (Hard)

Write a program to check if a number is a perfect number.

A perfect number is a positive integer that is equal to the sum of its proper positive divisors (excluding the number itself).

Example: 6 is perfect because 6 = 1 + 2 + 3

Example: 28 is perfect because 28 = 1 + 2 + 4 + 7 + 14

**Hint:** Loop from 1 to n-1. If i divides n (n%i==0), add i to sum. Compare sum with n.

````python
while True:
    try:
        n = int(input('Enter a number: '))
        break
    except ValueError:
        print('Enter a valid number.')
        
total = 0
for i in range(1, n):
    if n%i == 0:
        total += i
if n == total:
    print('Perfect number')
else:
    print('Not perfect number')
````

##### Problem 6.4: LCM Calculator (Medium)

Write a program to find the Least Common Multiple (LCM) of two numbers.

Hint: LCM(a, b) = (a × b) / GCD(a, b)

You can use the GCD function from Problem 6.2.

**Hint:** First calculate GCD using Euclidean algorithm (Problem 6.2), then use formula: LCM = (a * b) // GCD.

````python
import math
while True:
    try:
        a = int(input('Enter 1st number: '))
        b = int(input('Enter 2nd number: '))
        break
    except ValueError:
        print('Enter a valid number.')
lcm = (a*b)//math.gcd(a,b)
print(lcm)
````

##### Problem 6.5: Number to Words (Hard)

Write a program that converts a number (0-999) to words.

Examples:
- 0: Zero
- 25: Twenty Five
- 123: One Hundred Twenty Three
- 500: Five Hundred

**Hint:** Create lists for ones, teens, tens, etc. Extract hundreds, tens, ones digits. Convert each part to words and combine.

````python
while True:
    try:
        n = int(input('Enter a number: '))
        break
    except ValueError:
        print('Enter a valid number.')

ones = ['', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']
teens = ['ten', 'eleven', 'twelve', 'thirteen', 'fourteen', 'fifteen', 'sixteen', 'seventeen', 'eighteen', 'ninteen']
tens = ['', '', 'twenty', 'thirty', 'fourty', 'fifty', 'sisty', 'seventy', 'eighty', 'ninty']

if n == 0:
    print('Zero')
else:
    words = ''  
    hundreds = n//100
    remainder = n%100
    
    if hundreds > 0:
        words += ones[hundreds] + ' Hundred '
        
    if 10 <= remainder <= 19:
        words += teens[remainder - 10]
    else:
        words += tens[remainder // 10] + ' '
        words += ones[remainder % 10]
        
    print(words.title())
````

##### Problem 6.6: Strong Number Checker (Hard)

Write a program to check if a number is a Strong number.

A Strong number is a number whose sum of factorial of digits is equal to the original number.

Example: 145 is a strong number because 1! + 4! + 5! = 1 + 24 + 120 = 145

**Hint:** Extract each digit, calculate its factorial, add to sum. Compare sum with original number.

````python
while True:
    try:
        n = int(input('Enter a number: '))
        break
    except ValueError:
        print('Enter a valid number.')
o = n
total = 0
for i in str(n):
    fact = 1
    for j in range(1, int(i)+1):
        fact *= j
    total += fact
if total == o:
    print('Strong number')
else:
    print('Not a Strong number')  
````

##### Problem 6.8: Simple Calculator (Medium)

Create a simple calculator program that:
1. Shows a menu of operations: Add, Subtract, Multiply, Divide, Exit
2. Takes two numbers as input
3. Performs the selected operation
4. Shows the result
5. Repeats until the user chooses to exit
6. Handles division by zero error

**Hint:** Use while loop. Display menu, get choice. Use if-elif to perform operation. Check if divisor is 0 before division.

````python
print('Simple Calculator')
print('1. Add')
print('2. Subtract')
print('3. Multiply')
print('4. Divide')
print('5. Exit')
while True:
    try:
        choice = int(input('Enter your Choice:'))
    except ValueError:
        print('Invalid Choice! Please enter a number between 1 and 5')
        continue
    
    if choice == 5:
        print('Exiting Calculator..')
        break
    
    try:
        n1 = int(input('Enter your first number'))
        n2 = int(input('Enter your second number'))
    except ValueError:
        print('Enter a valid number')
        continue
    
    if choice == 1:
        print(f'Result = {n1}+{n2} = {n1 + n2}')
    elif choice == 2:
        print(f'Result = {n1}-{n2} = {n1 - n2}')
    elif choice == 3:
        print(f'Result = {n1}x{n2} = {n1 * n2}')
    elif choice == 4:
        if n2 == 0:
            print('Error: Division by zero is not allowed')
        else:
            print(f'Result = {n1}/{n2} = {n1 / n2}')
````
