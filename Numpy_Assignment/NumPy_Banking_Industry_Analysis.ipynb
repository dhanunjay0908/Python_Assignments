#### Question 15: Advanced Array Operations

```python
# **Q15.1:** Apply a function to calculate 5% interest on all account balances

def interest_balance(x):
    return x+x*0.05

print(interest_balance(banking_data[:, 1]))
```

```python
# **Q15.2:** Use `np.where` to categorize customers by balance ranges

balence = banking_data[:,1]
categorize = np.where(balence < 5000, "Low Balence", 
                      np.where(balence < 10000, 'Medium Balence', 'High Balence'))
print(categorize)
```

```python
# **Q15.3:** Use `np.select` to categorize customers by credit score ranges

credit_score = banking_data[:, 2]
condition = [credit_score < 250, (credit_score >= 550) & (credit_score < 700), credit_score >= 700]
choices = ["Credit score is low", "Credit score is medium", "Credit score is high"]
categorize = np.select(condition, choices, default="Unknown")
print(categorize)
```

```python
# **Q15.4:** Use `np.piecewise` to apply different interest rates based on balance

def low_intrest(x): return x+x*0.3
def medium_intrest(x): return x+x*0.5
def high_intrest(x): return x+x*0.8
# np.piecewise(array, conditions, functions)
intrest_per_balance = np.piecewise(
    banking_data[:, 1],
    [banking_data[:, 1] < 6000, (banking_data[:, 1] >= 6000) & (banking_data[:, 1] < 12000), banking_data[:, 1] >= 12000],
    [low_intrest, medium_intrest, high_intrest]
)
print(intrest_per_balance)
```

```python
# **Q15.5:** Use `np.apply_along_axis` to calculate the sum of each row

# np.apply_along_axis(function, axis, array)
sum_of_row = np.apply_along_axis(np.sum, 1, banking_data)
print(sum_of_row)
```
